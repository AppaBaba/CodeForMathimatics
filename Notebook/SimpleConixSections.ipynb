# Conic Sections

## General Equation

 $ax^2 + bxy + cy^2 + dx + ey + f = 0$
 
 identify the type of conic that is represented by a given equation

 If $b^2 − 4ac$ is less than zero, if a conic exists, it will be either a circle or an ellipse.
 
 If $b^2 − 4ac$ equals zero, if a conic exists, it will be a parabola.
 
 If $b^2 − 4ac$ is greater than zero, if a conic exists, it will be a hyperbola.

```python
from numpy import linspace, meshgrid
from pylab import axhline, axvline, contour, show
```

```python
x = linspace(-9, 9, 400)
y = linspace(-5, 5, 400)
x, y = meshgrid(x, y)
```

```python
def axes():
    axhline(0, alpha=.2, c = 'c')
    axvline(0, alpha=.2, c = 'c')
```

# Circle

 ## Standar Form of Circle Equation

Circle: $(x - h)^2 + (y - k)^2 = r^2$  Center(h, k) and Radius is r

```python
a = 4
b = a
h = 1
k = 1
axes()
mycircle = (x - h)**2/a**2 + (y - k)**2/b**2
contour(x, y, mycircle, [1], colors='r')
show()
```

# Ellipse

 ## Standar Forms of Ellipe Equation

Ellipse with horizontal major axis:

$(x - h)^2/a^2 + (y - k)^2/b^2 = 1$

```python
a = 5
b = 3
h = 2
k = 1
axes()
myellipse = (x - h)**2/a**2 + (y - k)**2/b**2
contour(x, y, myellipse, [1], colors='g')
show()
```

Ellipse with vertical major axis:

$(x - h)^2/b^2 + (y - k)^2/a^2 = 1$

```python
a = 4
b = 2
h = 1
k = 1
axes()
myellipse = (x - h)**2/b**2 + (y - k)**2/a**2
contour(x, y, myellipse, [1], colors='g')
show()
```

 ## Non-standar Forms of Ellipe Equation
 
$ax^2 + bxy + cy^2 + dx + ey + f = 0$

$b^2 -ac < 0$

```python
a = 2
b = -3
c = 2
d = 6
e = -5
f = -5
assert b**2 - 4*a*c < 0
axes()
nsellipe = a*x**2 + b*x*y + c*y**2 + d*x + e*y + f
contour(x, y, nsellipe, [0], colors='g')
show()
```

# Parabola
## Standar Form

$y^2 = 4ax$, with a < 0

```python
a = .4
axes()
myparabola = y**2 - 4*a*x
contour(x, y, myparabola, [0], colors='b')
show()
```

```python
a = -.4
axes()
myparabola = y**2 - 4*a*x
contour(x, y, myparabola, [0], colors='b')
show()
```

## Non-standar Form of Parabola Equation

$ax^2 + bxy + cy^2 + dx + ey + f = 0$

$b^2 − 4ac$ = 0

```python
a = 1
b = -2
c = 1
d = 3
e = 3
f = -9
assert b**2 - 4*a*c == 0
axes()
nsparabola = a*x**2 + b*x*y + c*y**2 + d*x + e*y + f
contour(x, y, nsparabola, [0], colors='b')
show()
```

# Hyperbola

## Standar Form

 $y^2/a^2 - x^1/b^2 = 1$

```python
a = 2
b = 1
axes()
myhyperbola = (x**2/a**2 - y**2/b**2)
contour(x, y,myhyperbola, [1], colors='m')
show()
```
